# Replace values in columns:

### map: will change absent values in the dictionary to Nan

### replace: won't do any thing

```python
d = {'No':False,'Yes':True}
df['International plan'] = df['International plan'].map(d)
df.head()
```

```python
df = df.replace({'Voice mail plan':d})
df.head()
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

## crosstab: contingency table between two variables

```python
pd.crosstab(df['Churn'],df['International plan'])
```

```python
pd.crosstab(df['Churn'],df['Voice mail plan'],normalize=True)
```

## df.pivot_table, takes three arguments:

values  a list of variables to calculate statistics for

index  a list of variables to group data ,

aggfunc  what statistics we need to calculate for groups, e.g. sum, mean, maximum, minimum or something ee.

e.



```python
df.pivot_table(["Total day calls", "Total eve calls", "Total night calls"],["Area code"],aggfunc='mean')
```

```python
total_calls = (df["Total day calls"]+df["Total eve calls"]+df["Total night calls"]+df["Total intl calls"])

# loc parameter is the number of columns after which to insert the Series object
# we set it to len(df.columns) to paste it at the very end of the dataframe
df.insert(loc=len(df.columns),column='Total calls',value=total_calls)
df.head()
```

```python
# could done be easier without a Series instance
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
```

```python
# delete columns
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
#delete rows
df.drop([1, 2]).head()
```

```python
pd.crosstab(df['Churn'],df['International plan'],margins=True)
```
